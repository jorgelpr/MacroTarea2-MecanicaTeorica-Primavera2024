Se obtuvo en el PDF que el potencial efectivo está dado por

$$V_{ef}(r) = \dfrac{\ell^2}{2\mu r^2} - \dfrac{k}{r} e^{\textstyle -\frac{r}{a}}$$

donde, para el potencial de Yukawa, la constante $k$ representa la fuerza de la interacción entre ambas partículas, y está dada por

$$k= k_C q_1 q_2$$

siendo $k_C \approx 8.9875 \times 10^{9} \frac{Nm^2}{C^2}$ es la constante de Coulomb, y $q_1, q_2$ son las cargas eléctricas de las partículas $m_1$ y $m_2$ respectivamente.

Se considera el escenario de que las dos partículas sean el deuterón $_{1}^{2}H^{+}$ y el tritón $_{1}^{3}H^{+} $, los cuales tienen masas $m_{_{1}^{2}H^{+}}=2 \,u$ y $m_{_{1}^{3}H^{+}}=3 \,u$, donde $u$ es la unidad de masa atómica tal que $u\approx 1.6605 \times 10^{-27} \, kg$

La masa reducida del sistema está dada por

$$\mu = 1.9926 \times 10^{-27}\, kg $$

Para este caso, $q_{_{1}^{2}H^{+}} = q_{_{1}^{3}H^{+}} \approx 1.602\times 10^{-19} \, C $

Por tanto, 

$$k = 2.3066 \times 10^{-28} Nm^2$$

```python
u = 1.6605e-27
m1 = 2*u
m2 = 3*u
mu = (m1*m2)/(m1+m2)

q1 = 1.602e-19
q2 = 1.602e-19
Coulomb_k = 8.9875e09
k= q1*q2*Coulomb_k
print(f'El valor de k es: {k:2.4e}')
print(f'La masa reducida es igual a {mu:2.4e} kg')
```

Se tomaron como valores representativos a $\ell_1 = 1.05 \times 10^{-34} \frac{kgm^2}{s}$ y  $\ell_2 = 2.5 \times 10^{-34} \frac{kgm^2}{s}$, y por indicación del problema, $a=1,2,3$

```python
l_i = [1.05e-34, 2.5e-34]
a_i = [1, 2, 3]
```

### Gráfico del potencial efectivo


```python
r = np.linspace(1e-15, 1e-14, 100)

fig = plt.figure(figsize=(14, 8))
ax = fig.add_subplot(111)

for l in l_i:
    for a in a_i:
        V_eff = (l**2 / (2 * mu * r**2)) - (k / r) * np.exp(-r / a)
        ax.plot(r, V_eff, label=f'$\ell$={l:.1e}, $a$={a}')

plt.title('Potencial efectivo $V_{\text{eff}}$ para diferentes $\ell$ y $a$')
#plt.xlim(1e-15,1e-14)
#plt.ylim(1e-15,1e-12)
plt.xlabel('r')
plt.ylabel('$V_{eff}$')
plt.legend()
plt.grid(True)
plt.show()
```
